```python
import sys
sys.path.append('..')

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import Normalize
```

```python
# read umap.npy
umap = np.load("../data/umap.npy")
points = umap[:, :2]
colors = umap[:, 2:]
umap.shape, points.shape, colors.shape
```

```python
norm = Normalize()
normalized_embeddings = norm(points)
```

```python
cmap = plt.get_cmap('plasma')
colors = cmap(normalized_embeddings).mean(axis=1)
```

```python
colors.shape
```

```python
plt.scatter(normalized_embeddings[:, 0] * 2 - 1, normalized_embeddings[:, 1] * 2 - 1, color=colors, s=0.25)
```
